```python
#Loading DataSets 

import pandas as pd 


stocks = pd.read_csv("DataSets/Stock_Prices.csv")
portfolio = pd.read_csv("DataSets/Portfolio.csv")

stocks["Date"] = pd.to_datetime(stocks["Date"])
stocks.head()

```

```python
#Stock Trend Visualization 

import matplotlib.pyplot as plt

for stock in stocks["Stock"].unique():
    data = stocks[stocks["Stock"] == stock]
    plt.plot(data["Date"], data["Close"], label=stock)

plt.legend()
plt.title("Stock Trends")
plt.show() 
```

```python
#Portfolio Return

prices = stocks.pivot(index="Date", columns="Stock", values="Close")
returns = prices.pct_change()

portfolio_return = returns.mul(
    portfolio.set_index("Stock")["Weight"], axis=1
).sum(axis=1)

portfolio_return.mean()
```
